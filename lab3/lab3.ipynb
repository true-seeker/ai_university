```python
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

1. Загрузите набор данных. Учитывайте, что данные сохранены в кодировке Windows (cp1251)

```python
df = pd.read_csv("../data/olimpics_results.csv", encoding='cp1251', sep=';', header=0)
df
```

2. Заполните пропущенные значения в результатах решения задач. Каким значением лучше заполнять пропущенные значения (исходите из специфики набора данных)?

```python
df[['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']] = df[['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']].fillna(0)
df
```

3. Проанализируйте суммарные баллы по каждой задаче. Отсортируйте задачи по убыванию суммарных баллов и постройте столбиковую диаграмму, на которой задачи будут расположены в порядке убывания суммарных баллов.

```python
sorted_task_sums = df[['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']].sum().sort_values(ascending=False)
sorted_task_sums
```

```python
_, ax = plt.subplots()
ax.bar(sorted_task_sums.index, sorted_task_sums)

plt.show()
```
